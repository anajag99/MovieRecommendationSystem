In this code, we aim to rank movies based on a **weighted rating system** that balances both popularity (number of votes) and quality (average rating). The first step is to calculate **C**, the overall mean vote average of all movies in the dataset. This serves as a baseline for adjusting ratings. Next, we determine **m**, the minimum number of votes required for a movie to be considered. This threshold is set at the **90th percentile** of vote counts, ensuring that only the most popular movies are ranked. Movies with fewer votes than **m** are filtered out to prevent unfair ranking due to limited reviews.
In below formula
where **v** is the number of votes a movie has received, **R** is its average rating, **m** is the minimum votes required, and **C** is the mean rating across all movies. This formula ensures that movies with a high number of votes get more influence, while those with fewer votes are adjusted toward the mean. The function is applied to the filtered dataset, creating a new **weighted_rating** clumn.


```python
# Calculate overall mean vote average (C)
C = df['vote_average'].mean()

# Define m as the 90th percentile of vote_count
m = df['vote_count'].quantile(0.90)

# Function to compute the weighted rating
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m)) * R + (m/(v+m)) * C

# Apply weighted rating to all movies, filling NaN values for those with low vote_count
df['weighted_rating'] = df.apply(lambda x: weighted_rating(x) if x['vote_count'] >= m else np.nan, axis=1)

# Sorting movies by weighted rating (excluding NaN values)
demographic_recommendations = df.dropna(subset=['weighted_rating']).sort_values('weighted_rating', ascending=False)

# Display top 10 movies based on weighted rating
print(demographic_recommendations[['original_title', 'vote_average', 'vote_count', 'weighted_rating']].head(10))
```
